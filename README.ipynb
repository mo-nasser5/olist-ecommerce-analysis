# 📊 Olist E-Commerce Data Analysis

## 📌 Project Overview

This project presents a comprehensive analysis of the Brazilian Olist E-Commerce dataset. The objective is to uncover business insights related to sales performance, customer behavior, product categories, delivery efficiency, and customer satisfaction.

Through data cleaning, exploratory data analysis (EDA), and interactive dashboard development, the project provides actionable recommendations to support data-driven decision-making.

---

## 🎯 Business Objectives

* Analyze overall sales performance
* Identify top-performing product categories
* Understand customer purchasing behavior
* Evaluate delivery performance
* Analyze customer satisfaction and reviews
* Generate business recommendations

The datasets were cleaned, transformed, and merged to perform the analysis.

## 💡 Key Insights

* Top product categories generated the majority of revenue.
* Customer satisfaction is highly influenced by delivery performance.
* Sales showed seasonal fluctuations throughout the year.
* Delayed deliveries often resulted in lower review scores.
* A small number of categories contributed significantly to total sales.

---

## 🚀 Business Recommendations

* Improve logistics efficiency to reduce delivery delays.
* Invest more in high-performing product categories.
* Enhance customer experience initiatives.
* Develop retention strategies for repeat customers.
* Monitor customer feedback continuously.
